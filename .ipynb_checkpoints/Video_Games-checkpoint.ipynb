# Video Game Sales Analysis 

To analyse the data from the table and perform data visualizations in order to understand our problem statement better and provide with appropriate solutions. 

```python
import numpy as np                
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
vg = pd.read_csv('salesdata.csv')  # Reading the csv file using pandas library
```

```python
a = vg.head(10)         
print(a)
```

```python
vg.dtypes
```

```python
t=vg.head(20)
i=vg.head(100)    # Sampling data
```

```python
sales=a.Global_Sales      # Initializing sales variable as Global sales column from the dataframe
pb=a.Publisher            # Initializing pb variable as Publisher column from the dataframe
z=(a.Year)               # Initializing z variable as Year column from the dataframe
sales
          
```

```python
# Creating a bar graph, Top 10 Games Vs Global Sales

plt.figure(figsize=(15,5))   # defining figure size
plt.xticks(rotation = 35)    # rotating the names so they can fit in the image
plt.title('Top 10 performing Games')     # adding chart title
sns.barplot(x="Name",y="Global_Sales", hue="Year", data=a, width=0.65)  # using seaborn library to plot the chart
```

```python
### Limitation: Can't display values 
```

```python
# To determine the Global Market share of top 10 games by Publisher 
plt.title('Global Market Share by Publisher')
plt.bar('Publisher', 'Global_Sales' ,data=a)
```

#### Nintendo controls the Global Market Share of the top 10 video games. i.e. All the top 10 video games are published by Nintendo

```python
# The same can be represented by a pie chart as the values are in percentage
plt.title('Global Market Share by Publisher')
plt.pie(sales, labels = pb, colors='r', autopct='%1.1f%%') 

# What the library matplotlib does is that it performs the complex functions easily it adds up the global sales of top 10 games 
# and then individually calculates the percentages of each value.
```

```python
# Scatter Plot to determine which Publisher has the games on which Platform 
plt.title('Publisher Vs Platform')
plt.scatter('Publisher', 'Platform', data=t)
```

```python
vgf = pd.read_csv('video_game_films.csv')     # Reading the second table data
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(15,5))
plt.xticks(rotation = 35)
plt.title('Publisher and their Genres')
plt.scatter('Publisher', 'Genre', data=i, color='orange')
```

```python
vgf.head()
```

```python
vgf.dtypes
```

```python
vgf["Box Office(Million USD)"] = 0.0    # Adding a new column
vgf['Worldwide_box_office'] = vgf['Worldwide_box_office'].astype(float)    # Changing the data type to float
vgf["Box Office(Million USD)"] = vgf["Worldwide_box_office"]/1000000       # Diving the amount by a million
vgf['Box Office(Million USD)'] = vgf['Box Office(Million USD)'].round(decimals = 3)  # Rounding the final decimal values to 3 places
films=vgf.head(10)
films
```

```python
# Bar Graph to check the highest grossing movies 
plt.figure(figsize=(15,7))
plt.xticks(rotation = 35)
plt.title('Highest grossing movies')
sns.barplot(x="Name",y="Box Office(Million USD)", hue="Year", data=films, width=0.65) 
```

```python
# Reading the CSV file extracted by joining two table in MySQL to determine intersecting Publishers in both tables
join = pd.read_csv('JoinFinal.csv')
```

```python
# Combined the sum of Global Sales of individual publisher using excel nad extracting Data from query in MySQL 
sp = pd.read_csv('SalesvsPublishercsv.csv')
```

```python
# Combined the sum of Box Office Collection of individual publisher using excel nad extracting Data from query in MySQL 
bp = pd.read_csv('PublishervsBoxOffice.csv')
```

```python
# Merged two data frames
jspmerge = join.merge(sp)
```

```python
jspmerge.head()
```

```python
# Merged all the three tables 
fullmerge = jspmerge.merge(bp)
```

```python
lily=fullmerge.Global_Sales
fullmergef=fullmerge.sort_values(by=['Worldwide_box_office'],ascending=False)
fullmergef
```

```python
fullmergef.dtypes
```

```python
# Plotting the scatter plot for films and their box office collection and Global Sales as games
plt.figure(figsize=(15,7))
plt.xticks(rotation = 35)
sns.set(style='whitegrid')
plt.annotate('Highest Grossing Movie', xy=('Capcom', 1400), xytext=('Capcom', 1000), weight='bold',
            ha= 'center', arrowprops=dict(facecolor='blue', headwidth=10, width=5, shrink=0.05))
plt.annotate('Bestseller Game', xy=('Nintendo', 100), xytext=('Nintendo', 400), weight='bold',
            ha= 'left', arrowprops=dict(facecolor='blue', headwidth=10, width=5, shrink=0.05))

sns.scatterplot(x='Publisher', y='Worldwide_box_office', data=fullmergef, hue=lily, palette="Set2", s=400)
```
